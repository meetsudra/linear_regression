```python
# LABEL ENCODING 
# Label encoding is a technique used in machine learning and data preprocessing to 
# convert categorical (non-numeric) labels into numerical values.

import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("CarPrice_Assignment.csv")
cat_data = df.select_dtypes(include=['object']).columns
# Create a LabelEncoder instance
label_encoder = LabelEncoder()
# Iterate through the categorical columns and label encode them
for column in cat_data:
    df[column] = label_encoder.fit_transform(df[column])
# Now, your DataFrame 'df' contains label encoded values in the categorical columns
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop(['price'],axis = 1))
scaled_features = scaler.transform(df.drop('price',axis = 1))
```

```python
X = scaled_features
Y = df['price']
X= pd.DataFrame(data=X,columns = df.drop(columns=['price']).columns)
```

```python
X.corr()
```

```python
#importing varience inflation factor from the stats model
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X
## calculating VIF for every column
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(data.values,i)for i in range(data.shape[1])],index=data.columns)
    if vif.max()>5:
        print(vif[vif == vif.max()].index[0],'has been removed')
        data = data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print("No multicollinearity present anymore")
        return data
```

```python
for i in range(10):
    vif_data=MC_remover(vif_data)
vif_data.head()
```

```python
from sklearn.model_selection import train_test_split
X = vif_data
Y = df['price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=44)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
reg.coef_
```

```python
pred = reg.predict(X_test)
pred
```

```python
import statsmodels.api as sm
import numpy as np
from sklearn import metrics 

r2  = reg.score(X_test, Y_test)
MAE2 = metrics.mean_absolute_error(Y_test,pred)
MSE2 = metrics.mean_squared_error(Y_test,pred) 
RMSE2= (np.sqrt(metrics.mean_squared_error(Y_test, pred)))
MAPE2=metrics.mean_absolute_percentage_error(Y_test,pred)
VIF= 1/(1-r2)



print('r-squared: {}'.format(r2))
print('VIF: {}'.format(VIF))
print('---------------------------------------')
print('Mean absolute error: {}'.format(MAE2))
print('Mean squared error : {}'.format(MSE2))
print('Root mean squared error: {}'.format(RMSE2))
print('Mean  Absolute %  error: {}'.format(MAPE2))
```
